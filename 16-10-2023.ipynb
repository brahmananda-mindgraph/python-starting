```python
num1 = 4
num2 = 6

x = np.lcm(num1, num2)

print(x)
```

```python
import numpy as np

arr = np.array([3, 6, 9])

x = np.lcm.reduce(arr)

print(x)
```

```python
import numpy as np

num1 = 6
num2 = 9

x = np.gcd(num1, num2)

print(x)
```

```python
import numpy as np

arr = np.array([20, 8, 32, 36, 16])

x = np.gcd.reduce(arr)

print(x)
```
